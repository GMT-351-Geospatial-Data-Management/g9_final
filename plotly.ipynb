```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("charging_stations.csv")
```

```python
clipped_data = data[["full_id", "osm_id", "Y", "X"]]
```

```python
clipped_data["status"] = 0.0
clipped_data["power"] = 0.0
clipped_data["fast_charging"] = 0.0
clipped_data["cost"] = 0.0
clipped_data["brand"] = 0.0
clipped_data["level"] = 0.0
```

```python
status = []
for i in range(1685):
    if np.random.randint(2)==0:
        status.append("available")
    else:
        status.append("unavailable")
```

```python
capacity = []
for i in range(1685):
    capacity.append(np.random.randint(3, 10))
```

```python
fast_charging = []
for i in range(1685):
    if np.random.randint(2)==0:
        fast_charging.append(True)
    else:
        fast_charging.append(False)
```

```python
level = []

for i in range(1685):
    if np.random.randint(2)==0:
        level.append("Level 1")
    elif np.random.randint(2)==1:
        level.append("Level 2")
    else:
        level.append("Level 3")
```

```python
power = []

for i in range(1685):
    if level[i]=="Level 1":
        power.append(120)
    elif level[i]=="Level 2":
        power.append(np.random.randint(208, 240))
    else:
        power.append(480)
```

```python
charging_brands = ["Begum Energy", "Eticanium Energy", "S4M3T Energy", "Tosuns Energy"]

brand = []

for i in range(1685):
    
    rand_index = np.random.randint(4)
    brand.append(charging_brands[rand_index])
```

```python
cost = []

for i in range(1685):
    
    cost.append(round(np.random.uniform(0.2, 0.7), 2))
```

```python
payment_method = []

methods = ["Cash", "Credit Card", "Debit Card", "Others"]

for i in range(1685):
    rand_index = np.random.randint(4)
    payment_method.append(methods[rand_index])
```

```python
clipped_data["status"] = status
clipped_data["power"] = power
clipped_data["fast_charging"] = fast_charging
clipped_data["cost"] = cost
clipped_data["brand"] = brand
clipped_data["level"] = level
```

```python
gdf = gpd.GeoDataFrame(
    clipped_data, geometry=gpd.points_from_xy(clipped_data.Y, clipped_data.X), crs="EPSG:3857")
```
